```python
print("----- Student Marks Analyzer -----")

marks = {
    "Alice": [75, 80, 90],
    "Bob": [60, 70, 65],
    "Charlie": [90, 95, 88]
}

# Subject Names
Sub = ["Math", "Science", "English"]

num_stud = len(marks)

print("_______ Student Marks Analyzer Results _______")
for name in marks:
    avg = sum(marks[name]) / 3
    print(f"{name}: Average: {avg:.1f}")

max_total = 0
topper = ""
for name in marks:
    total = sum(marks[name])
    if total > max_total:
        max_total = total
        topper = name
print(f"\nTopper: {topper} with {max_total} marks")

print("\n_____ Subject wise Average _____ ")
for i in range(len(Sub)):
    total = 0
    for name in marks:
        total += marks[name][i]
    sub_avg = total / num_stud
    print(f"{Sub[i]}: {sub_avg:.1f}")
```
